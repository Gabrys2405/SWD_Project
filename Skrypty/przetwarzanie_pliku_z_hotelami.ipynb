Plik przetwarza dane z pliku "Hotele SWD - oryg.csv", aby pozbyć się problemów z importem danych.

```python
data = HotelLoader().zaladuj_csv('../Dane/Hotele SWD - oryg.csv').jako_dataframe()
data
```

```python
# Usuń białe znaki z nazw kolumn
nowe_kolumny = []
for s in data.columns:
    nowe_kolumny.append(s.strip())
data.columns = nowe_kolumny
data
```

```python
# Odrzuć puste wiersze
do_odrzucenia = []
for i in range(data.values.shape[0]):
    pass
    if not isinstance(data.iloc[i]['Nazwa hotelu'], str):
        do_odrzucenia.append(i)
data.drop(do_odrzucenia, inplace=True)
data
```

```python
# Usuń białe znaki z treści
df_obj = data.select_dtypes(['object'])

data[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
data
```

```python
# W kolumnach, które powinny być numeryczne, zamień przecinki na kropki
kolumny_numeryczne = ["Opinia", "Opinia dla lokalizacji", "Cena za pobyt[zł]", "Odległość od centrum [km]", "Komfort"]
df_obj = data[kolumny_numeryczne].select_dtypes(['object'])

data[df_obj.columns] = df_obj.apply(lambda x: x.str.replace(',', '.'))
data
```

```python
# W kolumnach numerycznych zmień typ na float
kolumny_numeryczne_type_dict = {}
for k in kolumny_numeryczne:
    kolumny_numeryczne_type_dict[k] = 'float'

data = data.astype(kolumny_numeryczne_type_dict)
print(data.dtypes)
data
```

```python
# Kolumnę 'Bezpłatny parking' zamień na bool
kolumna = "Bezpłatny parking"

data[kolumna] = data[kolumna].map({'Tak': True, 'Nie': False, True: True, False: False})
data
```

```python
# Zapisz do nowego pliku
data.to_csv('..\Dane\Hotele SWD - przetworzone.csv', sep=';', index=False, decimal=',')
```
